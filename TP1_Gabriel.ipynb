```python
import unidecode
import string
import glob
```

```python
Global_words = ""
for filename in glob.glob('input/*txt'):
    print(filename)
    archive = open(filename, 'r')
    Global_words += archive.read()
    Global_words += " "
    archive.close()
print(Global_words)
```

```python
def clean_text(text):
    text = text.lower() # Minusculo
    #text.strip() # Remove espaços em branco
    text = unidecode.unidecode(text) # Remove acentos

    # Remove Pontuação
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # separar palavras
    text = text.split()

    # Criar vetor de palavras
    Vector = list(set(text)) # set remove duplicatas
    Vector.sort()
    
    return Vector
```

```python
def BagOfWords(text1, text2):
    vector_bow = []

    for word in text1:
        if word in text2:
            vector_bow.append("1")
        else:
            vector_bow.append("0")
        
    return vector_bow
```

```python
Vector_Global_words = clean_text(Global_words)
print(Vector_Global_words)
```

```python

for filename in glob.glob('input/*txt'):
    archive = open(filename, 'r')
    content = archive.read()
    archive.close()
    content = clean_text(content)
    Bow = BagOfWords(Vector_Global_words, content)


    archive = open('output/BagOfWords.txt', 'w')
    archive.write(filename + ": " + str(Bow) + "\n")
    archive.close()

    print(filename + ": " + str(Bow))


content = ""
```
