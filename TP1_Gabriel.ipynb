```python
import unidecode
import string
```

```python
archive = open('input/entry1.txt', 'r')
content = archive.read()
archive.close()
print(content)
```

```python
def clean_text(text):
    text = text.lower() # Minusculo
    #text.strip() # Remove espaços em branco
    text = unidecode.unidecode(text) # Remove acentos

    # Remove Pontuação
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)

    # separar palavras
    text = text.split()

    # Criar vetor de palavras
    Vector = list(set(text))
    Vector.sort()
    
    return Vector
```

```python
text_formatted = clean_text(content)
print(text_formatted)
archive = open('output/text_formatted.txt', 'w')

for word in text_formatted:
    archive.write(word + '\n')
archive.close()
```

```python
archive = open('input/entry2.txt', 'r')
content2 = archive.read()
archive.close()

text_formatted2 = clean_text(content2)
print(text_formatted2)
```

```python
vector_bow = []

for word in text_formatted:
    if word in text_formatted2:
        vector_bow.append("1")
    else:
        vector_bow.append("0")
    
archive = open('output/BagOfWords.txt', 'w')

for word in vector_bow:
    archive.write(word)
archive.close()

print(vector_bow)
```
