```python
import unidecode
import string
```

```python
archive = open('input/entry.txt', 'r')
content = archive.read()
archive.close()
print(content)
```

```python
def clean_text(text):
    text =text.lower() # Minusculo
    text = unidecode.unidecode(text) # Remove acentos

    # Remove Pontuação
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator)
    return text
```

```python
text_formatted = clean_text(content)
Vector = text_formatted.split()
Vector = list(set(Vector))
Vector.sort()

archive = open('output/Vector.txt', 'w')

for word in Vector:
    archive.write(word + '\n')
archive.close()
```
